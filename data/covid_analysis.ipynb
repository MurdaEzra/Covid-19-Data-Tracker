```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
print(df.columns)
df.info()
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for specific countries
countries = ['India', 'USA', 'Kenya']
filtered_df = df[df['location'].isin(countries)]

# Line chart: Total cases over time
plt.figure(figsize=(10,6))
for country in countries:
    country_df = filtered_df[filtered_df['location'] == country]
    plt.plot(country_df['date'], country_df['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.show()
```

## Key Insight 1:
India and USA had similar trends in total cases initially, but diverged significantly post-2021.

## Key Insight 2:
Kenya showed a slower case growth and vaccination rollout compared to India and the USA.
